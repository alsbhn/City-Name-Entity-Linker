```python
with open ('C:/Users/alisobhani/Desktop/crime paper/city_crime/annot_data/train_data.json') as f:
    df = json.load(f)
```

```python
# get city description
city = pd.read_csv('C:/Users/alisobhani/Desktop/crime paper/city_crime/ner/city.csv')
city = city.to_dict(orient='record')
```

```python
# tag ner

import ast

not_tag = ['not_specific_to_city','not_sure','inter_national_politics','not_related_to_city','personal_announce','crime_incident','indirect_city_relation','seems_not_related_to_city','public_announce','local_politics','leisure_events','gossip_news','accident','bunch_of_new_announce','city_local_announce','spam','redundant','sport_news']

for news in df:
    news['ner']=1

for news in df:
    for tag in not_tag:
        if tag in news['labels']:
            news['ner']=0
            
for news in df:
    if ast.literal_eval(news['annot_city']) == []:
        news['ner']=0
```

```python
# add city descriptions
for i, news in enumerate(df):
    if ast.literal_eval(news['annot_city']) != [] and ast.literal_eval(news['annot_city'])!= ['San Angelo']:
        news['description']=[c['description'] for c in city if c['city']== ast.literal_eval(news['annot_city'])[0]][0]
```

```python
# shuffle data
from sklearn.utils import shuffle
df = pd.DataFrame(df)
df = shuffle(df)
df = df.to_dict(orient='record')
```

```python
# train-test split
df_train = df[:int(0.9*len(df))]
df_test = df[int(0.9*len(df)):]
```

```python
df_train_new = []
for news in df_train:
    if ast.literal_eval(news['annot_city']) != []:
        df_train_new.append({'id': news['id'],'url': news['url'],'text': news['text'],'title': news['title'],'dataset': news['dataset'],'labels': news['labels'],'annot_city': news['annot_city'],'summary_title': news['summary_title'],'summary_centroid': news['summary_centroid'],'ner': news['ner'],'description': news['description']})

    for c in city:
        if c['city'] not in ast.literal_eval(news['annot_city']):
            df_train_new.append({'id': news['id'],'url': news['url'],'text': news['text'],'title': news['title'],'dataset': news['dataset'],'labels': news['labels'],'annot_city': news['annot_city'],'summary_title': news['summary_title'],'summary_centroid': news['summary_centroid'],'ner': 0,'description': c['description']})
```

```python
# save data
with open ('C:/Users/alisobhani/Desktop/crime paper/city_crime/annot_data/train_data.json', 'w') as f:
    json.dump(df_train_new, f)
    
with open ('C:/Users/alisobhani/Desktop/crime paper/city_crime/annot_data/test_data.json', 'w') as f:
    json.dump(df_test, f)
```
